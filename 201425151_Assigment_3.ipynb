```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns

# Reading Excel file
file_path = 'Apple_Revenue_Data.xlsx'
df = pd.read_excel(file_path)

# Display first few rows of the dataset
print(df.head(15))
```

```python
# Calculate the count of 'Yes' and 'No' values in the 'Profit or not' column
profit_counts = df['Profit or not'].value_counts()

# Calculate the proportion of 'Yes' and 'No' values
profit_proportions = profit_counts / df['Profit or not'].count()

# Create a bar plot for 'Profit or not' column counts
fig, ax = plt.subplots()
sns.barplot(x=profit_counts.index, y=profit_counts.values, ax=ax)
ax.set_title("Count of 'Profit or not' Values")
ax.set_xlabel("Profit or not")
ax.set_ylabel("Count")
plt.show()

```

```python

# Create a bar plot for 'Profit or not' column proportions
fig, ax = plt.subplots()
sns.barplot(x=profit_proportions.index, y=profit_proportions.values, ax=ax)
ax.set_title("Proportion of 'Profit or not' Values")
ax.set_xlabel("Profit or not")
ax.set_ylabel("Proportion")
plt.show()
```

From the bar graphs presented, it can be observed that Apple Inc. experienced a greater number of non-profitable days compared to profitable days within the time frame analyzed. This finding suggests that, during the specified period, the company's stock exhibited a higher frequency of daily price declines, which resulted in negative returns for investors on those particular days. Further research is necessary to investigate the underlying factors contributing to this pattern and to assess the potential implications for Apple Inc.'s long-term performance and investor sentiment.

```python
# Numerical columns
numerical_columns = ['Day', 'Month', 'Year', 'Open', 'High', 'Low', 'Close', 'Volume']

# Calculate statistical values
mean_values = df[numerical_columns].mean()
median_values = df[numerical_columns].median()
mode_values = df[numerical_columns].mode().loc[0]
std_values = df[numerical_columns].std()
variance_values = df[numerical_columns].var()
range_values = df[numerical_columns].apply(lambda x: x.max() - x.min())
quartiles = df[numerical_columns].quantile([0.25, 0.5, 0.75]).T
iqr_values = quartiles.apply(lambda x: x.loc[0.75] - x.loc[0.25], axis=1)
skewness_values = df[numerical_columns].skew()
kurtosis_values = df[numerical_columns].kurtosis()

# Print the results
print("Mean values:\n", mean_values, "\n")
print("Median values:\n", median_values, "\n")
print("Mode values:\n", mode_values, "\n")
print("Standard deviation values:\n", std_values, "\n")
print("Variance values:\n", variance_values, "\n")
print("Range values:\n", range_values, "\n")
print("Quartiles:\n", quartiles, "\n")
print("Interquartile Range (IQR):\n", iqr_values, "\n")
print("Skewness values:\n", skewness_values, "\n")
print("Kurtosis values:\n", kurtosis_values, "\n")
```



Median:

For the 'Open', 'High', 'Low', and 'Close' columns, the median values represent the middle price points in each respective column. These values indicate that half of the price data points fall below or equal to the median value, while the other half fall above or equal to the median value. The median values for 'Open', 'High', 'Low', and 'Close' are 0.4002, 0.4066, 0.3930, and 0.3993, respectively.

Finally, the median value for the 'Volume' column, which is approximately 215,978,000, suggests that half of the trading volume data points are lower or equal to this value, while the other half are higher or equal to this value. This median volume provides a central reference point to understand the typical trading activity in the dataset.

Standard deviation:

The standard deviation values represent the dispersion or spread of data points in each column. The Day, Month, and Year columns have relatively smaller standard deviations, indicating that the data points are closer to their respective means. In contrast, the Open, High, Low, and Close columns have larger standard deviations, suggesting that the stock prices exhibit more variability over time. The Volume column has the largest standard deviation, which demonstrates that trading volumes can vary significantly and are more dispersed than the other numerical columns in the dataset.

Skewness:

For the 'Day' and 'Month' columns, the skewness values are 0.016329 and -0.009062, respectively. These values are close to 0, indicating that the distributions of days and months in the dataset are approximately symmetrical.

The 'Year' column has a skewness value of 0.003258, which is also close to 0, suggesting that the distribution of years in the dataset is nearly symmetrical.

The 'Open', 'High', 'Low', and 'Close' columns have skewness values of 3.059819, 3.062381, 3.057186, and 3.059840, respectively. These positive values indicate that the tails on the right side of the distributions are longer and fatter than the left side, implying that there are more extreme high values in these price columns compared to low values.

Lastly, the 'Volume' column has a skewness value of 2.394564, which is also positive. This indicates that the distribution of trading volume has a longer and fatter tail on the right side, meaning that there are more extreme high trading volume values compared to low trading volume values in the dataset.

Kurtosis Analysis: 

The kurtosis values for the Day, Month, and Year columns are negative, indicating that these distributions are platykurtic with lighter tails and fewer extreme values compared to a normal distribution. In contrast, the Open, High, Low, Close, and Volume columns exhibit positive kurtosis values, suggesting that these distributions are leptokurtic with higher peaks and heavier tails, containing more outliers and extreme values than a normal distribution. Overall, the data in the Day, Month, and Year columns are less prone to extreme values, while the Open, High, Low, Close, and Volume columns demonstrate a higher propensity for outliers.

```python
# Numerical columns
numerical_columns = ['Day', 'Month', 'Year', 'Open', 'High', 'Low', 'Close', 'Volume']

# Calculate statistical values
stats_df = pd.DataFrame(index=numerical_columns)
stats_df['mean'] = df[numerical_columns].mean()
stats_df['median'] = df[numerical_columns].median()
stats_df['mode'] = df[numerical_columns].mode().loc[0]
stats_df['std'] = df[numerical_columns].std()
stats_df['variance'] = df[numerical_columns].var()
stats_df['range'] = df[numerical_columns].apply(lambda x: x.max() - x.min())
stats_df['iqr'] = df[numerical_columns].quantile(0.75) - df[numerical_columns].quantile(0.25)
stats_df['skewness'] = df[numerical_columns].skew()
stats_df['kurtosis'] = df[numerical_columns].kurtosis()

# Plot statistical values
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(14, 20))
stats_to_plot = ['mean', 'median', 'mode', 'std', 'variance', 'range', 'iqr', 'skewness', 'kurtosis']

for ax, stat in zip(axes.flatten(), stats_to_plot):
    sns.barplot(x=numerical_columns, y=stats_df[stat], ax=ax)
    ax.set_title(f"{stat.capitalize()} Values")
    ax.set_ylabel(stat.capitalize())
    ax.set_ylim(bottom=0)

plt.tight_layout()
plt.show()
```

```python
# Perform basic statistical analysis on the 'Close' and 'Volume' columns
print("\nSummary statistics for 'Close' and 'Volume':\n", df[['Close', 'Volume']].describe())

# Plot the 'Close' column
fig, ax1 = plt.subplots()
ax1.plot(df.index, df['Close'], label='Close')
ax1.set_xlabel("Index")
ax1.set_ylabel("Close")
ax1.set_title("Close Prices")
plt.show()

# Plot the 'Volume' column
fig, ax2 = plt.subplots()
ax2.plot(df.index, df['Volume'], label='Volume', color='orange')
ax2.set_xlabel("Index")
ax2.set_ylabel("Volume")
ax2.set_title("Trade Volume")
plt.show()
```

### Statisctics for the 'Close' and 'Volume' columns
The summary statistics for the 'Close' and 'Volume' columns provide insights into the distribution and variability of the stock data. The average closing price is around 15.32, and the average daily trading volume is approximately 326,327,500 shares. The standard deviation values indicate dispersion in both closing prices and trading volumes. The dataset ranges from a minimum closing price of 0.0384 to a maximum of 181.2605, and a minimum trading volume of 1,001,504 shares to a maximum of 2,147,483,647 shares. The quartile values further illustrate the spread of the data, offering a comprehensive understanding of the stock's performance and trading activity during the analyzed period.
